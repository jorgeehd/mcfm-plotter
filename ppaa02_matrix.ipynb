```python

#Plotting distributions plots with comparisons as well as kfactors. 


var = ["m_diphoton" , "n_jets" , "pT_gamma_2" , 'pT_gamma_1' , "eta_gamma1" , "eta_gamma2" , "R_gamgam", "R_gamjet", "pT_jet_1" , "pT_jet_2", "cent" ]

order = ["NNLO" , "NLO" ,  "LO" ] 




for v in var: 
    data_orders = []
    for o in order: 
        print(f"For variable {v}, adding order {o}")
        data_orders +=  [dataset.Dataset.from_MATRIX_dir(matrix_dir= base_path , name = f"{v}_{o}" ,var=v, order=o , debug=False )]

            
    comp_ = compare.Compare(data_orders, name = f"comp-object")
    comp_.plot(logy = True , output_dir= f"{base_path}/custom_hists" , labels=["NNLO","NLO","LO"])
    if (var != 'cent'): 
        comp_.create_kfactor(num_dataset= comp_.datasets[1], denom_dataset= comp_.datasets[2] , name = "k_NLO" , new_y_axis_name=f"k_factor" , add=True )
        comp_.create_kfactor(num_dataset=comp_.datasets[0], denom_dataset= comp_.datasets[1], name = "k_NNLO", new_y_axis_name="k_factor" , add = True )

    

        #print(len(comp_.datasets))

        comp_.plot(logy= False, output_dir=f"{base_path}/custom_hists/kfactors", labels = [ "k_NLO", "k_NNLO"  ], sel = [0 ,0 ,0 , 1 , 1]) 
```
